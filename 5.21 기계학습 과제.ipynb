```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Parkinsson disease.csv')
```

```python
df.columns
```

```python
plt.scatter(df['HNR'], df['NHR'])
plt.show()
```

#### 선형회귀로 예측해보기

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

HNR = df[['HNR']]
NHR = df['NHR']
regr = LinearRegression()
regr.fit(HNR, NHR)
plt.scatter(HNR, NHR, alpha=0.7)
HNR_predict = regr.predict(HNR)
plt.plot(HNR, HNR_predict, color='r')
plt.show()
print('선형 회귀로 예측한 모델의 스코어 : {}'.format(regr.score(HNR, NHR).round(2)))
```

#### 다항 회귀 모델 

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(include_bias = False)
HNR_poly = poly.fit_transform(HNR)
regr = LinearRegression()
regr.fit(HNR_poly, NHR)
print('다항회귀 모델의 계수 = {}, 절편 = {}'.format(regr.coef_, regr.intercept_))
print('다항회귀로 예측한 모델의 스코어: {}'.format(regr.score(HNR_poly, NHR).round(2)))
```

```python
NHR_predict = regr.predict(HNR_poly)
plt.scatter(HNR, NHR)
plt.scatter(HNR, NHR_predict, color='r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

x = 8 * np.random.rand(100, 1) -4
y = 0.5 * x**3 +0.5 * x**2 + x +3

plt.scatter(x, y) # 기본 그래프

regr = LinearRegression() 
poly = PolynomialFeatures(include_bias = False)
x_poly = poly.fit_transform(x)
regr.fit(x_poly, y)
y_predict = regr.predict(x_poly)

print(x_poly.shape)
print(y_predict.shape)
plt.scatter(x, y_predict, color='r')
```
